```python
df.groupby('Item_Identifier')['Item_Weight'].mean()
```

```python
df['Item_Weight']=df.groupby('Item_Identifier')['Item_Weight'].transform(lambda x:x.fillna(x.mean()))
```

```python
df.isna().sum()
```

```python
df['Item_Weight'].fillna(df['Item_Weight'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
#for the outlet_size missing value lets calculate the mode:
df['Outlet_Size'].mode()
```

```python
#since mode value(medium) is the most frequant size in the dataframe I prefer the mode value to be as the placeholder for missing values in outlet_size
df['Outlet_Size']=df['Outlet_Size'].fillna('MISSING')
```

```python
#check
df.isna().sum()
```

```python
#check duplicates:
df.duplicated().sum()
```

```python
#we have inconsistant values in the Item_fat_content such as  , low fat , reg so we need to make the values like each other:
df['Item_Fat_Content']=df['Item_Fat_Content'].str.replace('reg','Regular')
```
