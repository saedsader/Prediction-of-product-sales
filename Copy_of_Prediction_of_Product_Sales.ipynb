#Lets start with modeling

```python
y=df2['Item_Outlet_Sales']
X=df2.drop(columns=['Item_Outlet_Sales','Item_Identifier','Outlet_Identifier'])
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
num_cols = X_train.select_dtypes("number").columns
print("Numeric Columns:", num_cols)
impute_mean = SimpleImputer(strategy='mean')
scaler = StandardScaler()
num_pipe = make_pipeline(impute_mean, scaler)
num_pipe
```

```python
num_tuple=('numeric',num_pipe,num_cols)
num_tuple
```

```python
ord_cols=['Item_Fat_Content','Outlet_Size']
```

```python
impute_na_ord = SimpleImputer(strategy='constant', fill_value='NA')
item_fat_order = ["NA", "Low", "Regular"]
outlet_size_order = ["NA", "Small", "Medium", "High"]
ordinal_category_orders = [item_fat_order,outlet_size_order]
ord_encoder = OrdinalEncoder(categories=ordinal_category_orders)
scaler_ord = StandardScaler()
ord_pipe = make_pipeline(impute_na_ord,ord_encoder,scaler_ord)
ord_pipe
```

```python
ord_tuple=('Ordinal',ord_pipe,ord_cols)
ord_tuple
```

```python
ohe_cols = X_train.select_dtypes('object').drop(columns=ord_cols).columns
print("OneHotEncoder Columns:", ohe_cols)
impute_na = SimpleImputer (strategy='constant', fill_value = "NA")
ohe_encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
ohe_pipe= make_pipeline(impute_na, ohe_encoder)
ohe_pipe
```

```python
ohe_tuple=('categorical',ohe_pipe,ohe_cols)
ohe_tuple
```

```python
col_transform=ColumnTransformer([num_tuple,ord_tuple,ohe_tuple],verbose_feature_names_out=False)
col_transform
```

```python
col_transform.fit(X_train)
X_train_tf=col_transform.transform(X_train)
X_test_tf=col_transform.transform(X_test)
```

```python
lin_reg=LinearRegression()
lin_reg
```

```python
lin_reg_f=lin_reg.fit(X_train_tf,y_train)
lin_reg_f
```

```python
def regression_metrics(y_true, y_pred, label='', verbose = True, output_dict=False):
  # Get metrics
  mae = mean_absolute_error(y_true, y_pred)
  mse = mean_squared_error(y_true, y_pred)
  rmse = np.sqrt(mse)
  r_squared = r2_score(y_true, y_pred)
  if verbose == True:
    # Print Result with Label and Header
    header = "-"*60
    print(header, f"Regression Metrics: {label}", header, sep='\n')
    print(f"- MAE = {mae:,.3f}")
    print(f"- MSE = {mse:,.3f}")
    print(f"- RMSE = {rmse:,.3f}")
    print(f"- R^2 = {r_squared:,.3f}")
  if output_dict == True:
      metrics = {'Label':label, 'MAE':mae,
                 'MSE':mse, 'RMSE':rmse, 'R^2':r_squared}
      return metrics

def evaluate_regression(reg, X_train, y_train, X_test, y_test, verbose = True,
                        output_frame=False):
  # Get predictions for training data
  y_train_pred = reg.predict(X_train)

  # Call the helper function to obtain regression metrics for training data
  results_train = regression_metrics(y_train, y_train_pred, verbose = verbose,
                                     output_dict=output_frame,
                                     label='Training Data')
  print()
  # Get predictions for test data
  y_test_pred = reg.predict(X_test)
  # Call the helper function to obtain regression metrics for test data
  results_test = regression_metrics(y_test, y_test_pred, verbose = verbose,
                                  output_dict=output_frame,
                                    label='Test Data' )

  # Store results in a dataframe if ouput_frame is True
  if output_frame:
    results_df2 = pd.DataFrame([results_train,results_test])
    # Set the label as the index
    results_df2 = results_df2.set_index('Label')
    # Set index.name to none to get a cleaner looking result
    results_df2.index.name=None
    # Return the dataframe
    return results_df2.round(3)
```

```python
evaluate_regression(lin_reg,X_train_tf,y_train,X_test_tf,y_test)
```

#Since train and test R² are almost equal, the model is not overfitting and generalizes well.

```python
rf=RandomForestRegressor(random_state=42)
rf_pipe=make_pipeline(col_transform,rf)
rf_pipe
```

```python
rf_pipe.fit(X_train,y_train)
```

```python
evaluate_regression(rf_pipe,X_train,y_train,X_test,y_test)
```

#The Random Forest model appears to be overfitting because the training R² score is high while the testing R² score is significantly lower. This means the model learned the training data very well, including noise and specific patterns, but did not generalize effectively to unseen data. Therefore, the model performs strongly on training data but weakly on new data.
To reduce overfitting in a Random Forest model, we can limit the complexity of the trees by reducing max_depth, increasing min_samples_leaf, and tuning hyperparameters using cross-validation. Additionally, feature selection and increasing training data can help improve generalization.

```python
rf_pipe.get_params()
```

```python
param={'randomforestregressor__max_depth': [None, 5,10,15],'randomforestregressor__min_samples_leaf': [2,5,10]}
```

```python
gridsearch_rf=GridSearchCV(rf_pipe,param,n_jobs=-1,cv=3,verbose=1)
```

```python
gridsearch_rf.fit(X_train,y_train)
```

```python
gridsearch_rf.best_params_
```

```python
best_rf=gridsearch_rf.best_estimator_
evaluate_regression(best_rf,X_train,y_train,X_test,y_test)
```

#After applying hyperparameter tuning, the training and testing R² scores became very close. This indicates that the model is more balanced and generalizes well to unseen data. It also suggests that the overfitting problem has been reduced and the model performance has improved.
